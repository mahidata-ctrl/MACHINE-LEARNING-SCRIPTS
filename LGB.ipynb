```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(labels="DEFAULT",axis=1)
```

```python
y=df["DEFAULT"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80)
```

```python
rf_model=RandomForestClassifier()
g_model=GradientBoostingClassifier()
ag_model=AdaBoostClassifier()
XGB_model=XGBClassifier()
LGB_model=LGBMClassifier()
```

```python
%%time
rf_model.fit(x_train,y_train)
```

```python
y_pred=rf_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%time
g_model.fit(x_train,y_train)
```

```python
y_pred=g_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%time
ag_model.fit(x_train,y_train)
```

```python
y_pred=ag_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%time
XGB_model.fit(x_train,y_train)
```

```python
y_pred=XGB_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%time
LGB_model.fit(x_train,y_train)
```

```python
y_pred=LGB_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
RF_feature=rf_model.feature_importances_
G_feature=g_model.feature_importances_
AG_feature=ag_model.feature_importances_
XGB_feature=XGB_model.feature_importances_
LGB_feature=LGB_model.feature_importances_
```
